Add the zip file location in the file section. Output folder will name as the extract_path variable.

```python
file = '/content/8ba813e8099e3a56e3d644a4aa76e87e.zip'
extract_path = 'tpT2mSstSp'
with zipfile.ZipFile(file, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
```

Inside the extract_path folder, select any one file location to put in the open_dataset function. Then, Show the data variable, to see which variables are inside the file.

```python
data = xr.open_dataset('/content/tpT2mSstSp/data_stream-oper_stepType-accum.nc')
data
```

Access Data Variable tp and show necessary informations.

```python
tp = data.tp
tp
```

```python
tp.shape
```

Convert tp variable to a one dimensional list by averaging the other dimensional values.

```python
tp_mean = []
for i in range(len(tp)):
  tp_mean.append(np.mean(tp[i]))
tp_mean=np.array(tp_mean)
tp_mean.shape
```

To understand data more accurately, convert this to a 2D data so that this could be accessable via month year index. Rows and columns represents months and years respectively. `tp_my[i, j]` means averavage precepitation for i month in j year.    

```python
month_length_n = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
month_length_lp = [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

tp_my = np.zeros((23, 12))
year = 2001
index = 0
start = 0
end = 0
for i in range (23):
  if year % 4 == 0:
    month_length = month_length_lp
  else:
    month_length = month_length_n

  for j in range (12):
    start = end
    end = start + month_length[j]
    tp_my[i, j] = np.mean(tp_mean[start:end])
  year += 1
tp_my.shape
```

Yearly average plotting to understand data trend

```python
x = [i for i in range (23)]
y = []
for i in range(23):
  y.append(np.mean(tp_my[i, :]))
y = np.array(y)
plt.bar(x, y)
plt.title('Yearly Average')
plt.ylabel('meter')
plt.xlabel('year')
plt.show()
print("Mean precipitation(2001-2023): ", np.mean(y), "m")
print("Mean precipitation(2001-2014): ", np.mean(y[:14]), "m")
print("Mean precipitation(2015-2023): ", np.mean(y[-9:]), "m")
print("Change of mean: ",np.mean(y[:14]) - np.mean(y[-9:]), "m")
```

Monthly average plotting to get the seasonal understanding for that particular place.

```python
x = [i for i in range (12)]
y = np.zeros([12])
for i in range(12):
  y[i] = np.mean(tp_my[:, i])
plt.bar(x, y)
plt.title('Monthly Average')
plt.ylabel('meter')
plt.xlabel('month')
plt.show()
```

According to the graph, November is the top and February is the bottom point.  

```python
fig, (g1, g2) = plt.subplots(1, 2, figsize = (14, 6))

x = [i for i in range (23)]
y = tp_my[:,9]
g1.bar(x, y)
g1.set_title('November(Maximum)')
g1.set_ylabel('meter')
g1.set_xlabel('year')

y = tp_my[:,1]
g2.bar(x, y)
g2.set_title('February(Minimum)')
g2.set_ylabel('meter')
g2.set_xlabel('year')
plt.show()
```

To check the monthly trend (if any pattern or shifting of the parameter)

```python
for i in range(0, 22):
  x = [i for i in range (12)]
  y = tp_my[i, :]
  plt.plot(x, y, 'bo-')
  plt.title('Monthly Average')
  plt.ylabel('meter')
  plt.xlabel('month')
  plt.show()
```

Monthly average check before and after to establishment.

```python
fig, (g1, g2) = plt.subplots(1, 2, figsize = (14, 6))

x = [i for i in range (12)]
y = np.zeros([12])
for i in range(12):
  y[i] = np.mean(tp_my[:14, i])
g1.bar(x, y)
g1.set_title('Monthly average of Year: 2001-2014')
g1.set_ylabel('meter')
g1.set_xlabel('month')

for i in range(12):
  y[i] = np.mean(tp_my[-9:, i])
g2.bar(x, y)
g2.set_title('Monthly average of Year: 2015-2023')
g2.set_ylabel('meter')
g2.set_xlabel('Month')
plt.show()
```

Final plotting of the comparison. Uncomment plt.savefig line to save the plot for further using.

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
x = np.arange(len(months))
y1 = []
y2 = []

for i in range(12):
  y1.append(np.mean(tp_my[:14, i]))
for i in range(12):
  y2.append(np.mean(tp_my[-9:, i]))


fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x , y1, marker='o', label='Before 2014', color='green')
ax.plot(x , y2, marker='o', label='After 2014', color='red')

ax.set_xlabel('Months')
ax.set_ylabel('Meter')
ax.set_title('Monthly average')
ax.set_xticks(x)
ax.set_xticklabels(months)
ax.legend()

plt.tight_layout()
#plt.savefig("tp", dpi = 600)
plt.show()
```
